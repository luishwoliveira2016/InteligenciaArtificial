
**Agora use as matrizes x, ye z para recriar o gráfico abaixo. Observe os limites xlimits e y no gráfico inserido:**

```python
x = np.linspace(0, 100, 10000)
y = x**2

fig = plt.figure()

axes1 = fig.add_axes([0, 0, 1, 1]) 
axes2 = fig.add_axes([0.2,0.5,.4,.4])

axes1.plot(x, y, 'b')
axes1.set_xlabel('x')
axes1.set_ylabel('y')

x = np.linspace(20, 22, 50)
y = np.linspace(10, 50, 50)

axes2.plot(x, y, 'b')
axes2.set_xlabel('x')
axes2.set_ylabel('y')
axes2.set_title('zoom');

plt.show()
```

## Exercício 4

**Use plt.subplots (nrows = 1, ncols = 2) para criar o gráfico abaixo.**

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

for ax in axes:
    ax.plot(x, y, 'g')
    ax.set_xlabel('x')
    ax.set_ylabel('y')

plt.tight_layout()
```

**Agora plote (x, y) e (x, z) nos eixos. Brinque com a largura de linha e o estilo**

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

for ax in axes:
    ax.plot(x, y, 'b--')
    ax.set_xlabel('x')
    ax.set_ylabel('y')

fig
plt.tight_layout()
```


**Veja se você pode redimensionar o gráfico adicionando o argumento figsize () em plt.subplots () está copiando e colando o código anterior.**

```python
fig, axes = plt.subplots(nrows=1, ncols=2,figsize=(20, 10))

for ax in axes:
    ax.plot(x, y, 'b--')
    ax.set_xlabel('x')
    ax.set_ylabel('y')

fig
plt.tight_layout()
```
